```python
# 导入相应模块
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cooked data and the corresponding path
# local_path = 'C:/Users/YangL'
#local_path = 'C:/Users/W/Python基础课件'
basedata_path = 'basedata.csv'
consensus_path =  'disclose.csv'
calendar_path =  'ashare_calendar.csv'
```

```python
# 1. import daily stock return and market return
basedata = pd.read_csv(basedata_path)
basedata['abret'] = basedata['ret'] - basedata['mkt_ret'] #这里以market return
basedata['date'] = pd.to_datetime(basedata['date'])

# 2. import analyst consensus and earning surprise 
consensus = pd.read_csv(consensus_path)
consensus['发布时间'] = pd.to_datetime(consensus['发布时间'])

# 3. import trading calendar for Ashare stocks
ashare_calendar = pd.read_csv(calendar_path)
ashare_calendar['Date'] = pd.to_datetime(ashare_calendar['Date'])
```

```python
# 基于公告日，建立事件研究时间窗口
event_day = ashare_calendar.copy()
for i in range(-15,16): #每个交易日的前后15个交易日
    event_day[i] = event_day['Date'].shift(-i)
    
## 计算每一个trading day 前后15个交易日对应的日期
eventdays = event_day.set_index('Date').stack().reset_index().rename(columns={'level_1':'event_day',0:'tdate'})

consensus2 = consensus[['股票代码', '发布时间', '预告类型']].sort_values(by=['股票代码', '发布时间'])
```

```python
# 1、找到每一个公告日前后15个交易日的时间窗口
all_info = pd.merge(consensus2, eventdays, left_on='发布时间', right_on='Date', how='inner')
all_info = pd.merge(all_info, basedata, left_on=['股票代码','tdate'],
                    right_on = ['stock_code','date'], how='inner')
all_info.sort_values(by=['stock_code','发布时间','event_day'],inplace=True)

result1 = all_info.groupby(['预告类型','event_day']).agg({'ret':'mean',}).reset_index()
result1['预告类型'] = result1['预告类型'].astype(str)
result1.sort_values(by=['预告类型','event_day'],inplace=True)

# 2、找到每一个公告日前后15个交易日的时间窗口
all_info = pd.merge(consensus2, eventdays, left_on='发布时间', right_on='Date', how='inner')
all_info = pd.merge(all_info, basedata, left_on=['股票代码','tdate'],
                    right_on = ['stock_code','date'], how='inner')
all_info.sort_values(by=['stock_code','发布时间','event_day'],inplace=True)

result2 = all_info.groupby(['预告类型','event_day']).agg({'abret':'mean',}).reset_index()
result2['预告类型'] = result1['预告类型'].astype(str)
result2.sort_values(by=['预告类型','event_day'],inplace=True)

# 3、找到每一个公告日前后15个交易日的时间窗口
all_info = pd.merge(consensus2, eventdays, left_on='发布时间', right_on='Date', how='inner')
all_info = pd.merge(all_info, basedata, left_on=['股票代码','tdate'],
                    right_on = ['stock_code','date'], how='inner')
all_info.sort_values(by=['stock_code','发布时间','event_day'],inplace=True)

result = all_info.groupby(['预告类型','event_day']).agg({'abret':'mean',}).reset_index()
result['预告类型'] = result['预告类型'].astype(str)
result.sort_values(by=['预告类型','event_day'],inplace=True)

# 从公告日前5个交易日开始画图
result = result.loc[result['event_day']>=-5]
result['ret2'] = result['abret'] + 1

# 计算累积异常收益率
result['cum_ret'] = result.groupby('预告类型')['ret2'].transform('cumprod')-1
data1=result[result['预告类型'].isin(['首亏','续亏','扭亏','预增','续盈','大增'])]
st1 = pd.merge(result1, result2, on=['预告类型','event_day'],how='inner')
```
