#### 不同数组的叠加

```python
import numpy as np
a1 = np.random.randint(0, 10, size = (3, 4))
a2 = np.random.randint(0, 10, size = (1, 4))
a3 = np.vstack([a1, a2])  # 垂直方向叠加
print(a1)
print('=' * 20)
print(a2)
print('=' * 20)
print(a3)
```

```python
import numpy as np
a1 = np.random.randint(0, 10, size = (3, 4))
a2 = np.random.randint(0, 10, size = (3, 1))
a3 = np.hstack([a1, a2])  # 水平方向叠加
print(a1)
print('=' * 20)
print(a2)
print('=' * 20)
print(a3)
```

```python
import numpy as np
a1 = np.random.randint(0, 10, size = (3, 3))
a2 = np.random.randint(0, 10, size = (3, 3))
a3 = np.concatenate((a1, a2), axis = 0) 
print(a1)
print('=' * 20)
print(a2)
print('=' * 20)
print(a3)
```

```python
import numpy as np
a1 = np.random.randint(0, 10, size = (3, 3))
a2 = np.random.randint(0, 10, size = (3, 3))
a3 = np.concatenate((a1, a2), axis = 1) 
print(a1)
print('=' * 20)
print(a2)
print('=' * 20)
print(a3)
```

```python
import numpy as np
a1 = np.random.randint(0, 10, size = (3, 3))
a2 = np.random.randint(0, 10, size = (3, 3))
a3 = np.concatenate((a1, a2), axis = None) 
print(a1)
print('=' * 20)
print(a2)
print('=' * 20)
print(a3)
```

#### 曼德勃罗分形图形

```python
import time
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
wid_size = 1920
hei_size = 1080
maxVal = 255
orig_x = wid_size * 2 / 3
orig_y = hei_size / 2
def iteration(x, y):
    limit = maxVal + 1
    A = (x - orig_x) / (wid_size / 3)
    B = (orig_y - y) / (hei_size / 2)
    c = complex(A, B)
    z = complex(0, 0)
    for i in range(limit):
        z = z * z + c
        if z.real > 2 or z.imag > 2:
            break
    return 0 if i == maxVal else 1
```

```python
if __name__ == '__main__':
    img = Image.new('RGB', (wid_size, hei_size))
    # img.shape
    imgp = np.array(img)
    t0 = time.time()
    for i in range(hei_size):
        for j in range(wid_size):
            ite = iteration(j ,i)
            if ite:
                imgp[i, j, 0] = ite
                imgp[i, j, 1] = abs(j - orig_x) / wid_size * maxVal
                imgp[i, j, 2] = abs(i - orig_y) / hei_size * maxVal
            else:
                imgp[i, j, :] = 128
                
    t1 = time.time()
    print(t1 - t0,'s')
    plt.figure(figsize = (25,10))
    plt.imshow(imgp)
    plt.show()
    img = Image.fromarray(imgp)
    # img.save('C:/Users/jiangyu/Desktop/1.jpg')
```
