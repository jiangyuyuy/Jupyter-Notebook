```python
import numpy as np
import matplotlib.pyplot as plt
a = np.array([0, 0])
b = np.array([1, 1])
dot_1 = []
for t in np.arange(0, 1, 0.1):
    c = a * t + b * t
    dot_1.append(c)
    
dot_1
```

```python
plt.figure(figsize=(6, 4), dpi=500)
plt.plot(dot_1, marker='*', linestyle='-.', c='r', linewidth=0.5)
plt.show()
```

```python
p0 = np.array([0, 0])
p1 = np.array([1, 2])
p2 = np.array([2, 2])
p3 = np.array([3, 0])

dot_4 = []
for t in np.arange(0, 1.05, 0.05):
    
    # 四个控制点的贝塞尔曲线
    pt = (1 - t)**3 * p0 + 3*t * (1 - t)**2 * p1 + 3 * t**2 * (1 - t) * p2 + t**3 * p3
    dot_4.append(list(pt))

y_value = []
x_value = []
for i in dot_4:
    x_value.append(i[0])
    y_value.append(i[1])
```

```python
plt.figure(figsize=(6, 4), dpi=500)
plt.plot([0, 1, 2, 3], [0, 2, 2, 0], linewidth=0.5, c='g')
plt.plot(x_value, y_value, marker='*', linestyle='-.', c='r', linewidth=0.5)
plt.show()
```
